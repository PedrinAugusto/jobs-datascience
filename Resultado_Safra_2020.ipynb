

- Apesar do agro gerar muito lucro, a vida dos agricultores não é fácil, mas sim um verdadeiro teste de resistência e determinação. Uma vez que tenhamos semeado as sementes, o agricultor precisa trabalhar dia e noite para garantir uma boa safra no final da estação. Uma boa colheita depende de diversos fatores, como disponibilidade de água, fertilidade do solo, proteção das culturas, uso oportuno de pesticidas, outros fatores químicos úteis e da natureza.

- Muitos desses dados são quase impossíveis de se controlar, mas a quantidade e a frequência de pesticidas é algo que o agricultor pode administrar. Os pesticidas podem protegem a colheita com a dosagem certa. Mas, se adicionados em quantidade inadequada, podem prejudicar toda a safra.

- Dito isto, abaixo são fornecidos dados baseados em culturas colhidas por vários agricultores no final da safra de 2018-2019. Para simplificar o problema, você pode assumir que todos os fatores relacionados as técnicas agrícolas e climáticas, não influenciaram esses resultados.

- Seu objetivo neste exercício é determinar o resultado desta safra atual de 2020, ou seja, se a colheita será saudável, prejudicada por pesticidas, ou prejudicada por outros motivos.

#  Informações dos dados: 

Nenhuma das colunas têm uma alta correlação com os dados causados na plantação. As maiores correlações são com a estimativa de incetos e o número de semanas utilizando.

- Como foi passado, pode-se assumir que todos os fatores relacionados as técnicas agrícolas e climáticas, não influenciarão esses resultados, além disso a coluna de identificador do agricultor também não agrega informação relevante para a análise. Dessa maneira, pode-se excluir as colunas 'Identificador_Agricultor', 'Tipo_de_Cultivo' e 'Temporada'.

- Todas as colunas possuem 80 mil dados, com excessão da coluna 'Semanas_Utilizando', que possui 71.945 dados, dessa forma é necessário uma análise mais aprofundada sobre esses dados faltantes. 

- Dividir os dados em dados de entrada e de resultado.

- As colunas 'Tipo_de_Solo', 'Categoria_Pesticida' e 'dano_na_plantacao' são dados categóricos, apesar de possuir números em sua classificação, esses números representam tipos e categorias e não um valor escalar de vantagem de um tipo ou categoria em relação à outra. Reorganizar essas informações. 

 


### 1º  Informações que não agregam para a análise.

```python
# Excluir colunas 
safra.drop(['Identificador_Agricultor','Tipo_de_Cultivo', 'Temporada'], inplace=True, axis=1)
safra.head()
```

### 2º Dados ausentes.

```python
# Pesquisa sobre dados únicos na coluna, afim de identificar a forma os dados os dados ausentes
safra['Semanas_Utilizando'].unique()
```

É possível identificar que todos os dados ausentes são NaN, além disso, é importante entender a qualidade desses dados para a análise, já que podemos substituir o valor desses NaN por algum outro valor (média por exemplo) ou podemos excluir esses dados da análise.
```python
safra_18_19 = safra.dropna() # Excluir todas as linhas com NaN
safra_18_19.describe() # Descrição dos dados
```

É possível observar que todas as colunas possuiem, agora, uma quantidade de 71945 dados.

### 3º Informações x Resultado

```python
# Separar os dados de informações, que geram as previsões (previ) e os dados de classificação (classe)
previ = safra_18_19.iloc[:, 0:6]
classe = safra_18_19.iloc[:, 6]
```

```python
previ.head() # Visualizar os dados de previsão atualizados
```

```python
classe.head() # Visualizar os dados de resultado
```

### 4º Organizar dados categóricos

```python
# Importar bibliotecas de processamento e de transformação
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# Criar o transformador OneHotEncoder para as informações 
column_tranformer = ColumnTransformer([('one_hot_encoder', OneHotEncoder(), [1, 2])],
                                      remainder='passthrough')
```

```python
# Transformação de dados categóriocos pelo OneHotEncoder
previ = column_tranformer.fit_transform(previ)
```

```python
pd.DataFrame(previ).head() # Visualização da atualização dos dados depois da transformação pelo OneHotEncoder
```

Agora são 9 colunas:
   - Duas colunas para os dados de tipo de solo 
   - Três colunas para os dados de categoria de pesticidas. 

Finalizada a sepearação dos dados e todos os ajustes, agora é só aplicar o método desejado. 
Como teste, usarei o Random Forest. 

```python
# Bibliotecas de separação dos dados de treinamento e de teste e do classificador Random Forest
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# Separeção de dados
previ_train, previ_test, classe_train, classe_test = train_test_split(previ, classe, test_size=0.15, random_state=0)
```

```python
# Criação do classificador random forest
classificador = RandomForestClassifier(n_estimators=50, criterion='entropy', random_state=0)
```

```python
# Passar dados para o classificador
classificador.fit(previ_train, classe_train)
```

```python
# Criar as previsões
previsoes = classificador.predict(previ_test)
previsoes
```

```python
# Bibliotecas de metricas da classificação 
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Precisão das previsões
precisao = accuracy_score(classe_test, previsoes)
print('O valor da precisão das previsões é de:',precisao)
```
